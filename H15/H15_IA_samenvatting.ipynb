# Onderwerp 11: Dictionary's - Samenvatting

<p>In een dictionary houd je paren van waarden bij, zoals <code>gamescores</code> hieronder.
Het linker deel heet de <b>sleutel</b> (Engels: key) en het rechter deel de <b>waarde</b>
(Engels: value).</p>

<img src="https://raw.githubusercontent.com/rweeda/PythonIA/main/img/dicitonary_gamescores.png" width="300">

```python
# 11.1: Dictionary maken en afdrukken
lege_dictionary = dict()

gamescores = {"Tim": 500, "Janneke": 900, "Bob": 680}
print(gamescores)  # dictionary afdrukken
```

```python
# 11.2: Waarde ophalen
gamescores = {"Tim": 500, "Janneke": 900, "Bob": 680}
score_van_bob = gamescores["Bob"]   # haalt de waarde van sleutel "Bob" op (680)

# Waarde ophalen met get()
score_van_janneke = gamescores.get("Janneke")  # haalt de waarde van sleutel "Janneke" op (900)
```

```python
# 11.3: Waarde toevoegen
gamescores = {"Tim": 500, "Janneke": 900, "Bob": 680}
gamescores["Karel"] = 750  # voegt een nieuw sleutelwaarde-paar to
```

```python
# 11.4: Waarde aanpassen
gamescores = {"Tim": 500, "Janneke": 900, "Bob": 680}
gamescores["Tim"] = 600  # verandert de waarde van sleutel "Tim" in 600
```

```python
# 11.6: Sleutelwaarde-paar verwijderen
gamescores = {"Tim": 500, "Janneke": 900, "Bob": 680}
del gamescores["Janneke"]  # verwijdert het sleutelwaarde-paar met sleutel "Janneke"
```

```python
# 11.7: Dictionary doorlopen met een for-loop
gamescores = {"Tim": 500, "Janneke": 900, "Bob": 680}
for sleutel in gamescores:
    print(sleutel)                     # print de sleutels van de dictionary
    print(gamescores[sleutel])         # print de bijbehorende waarde
    print(gamescores.get(sleutel))     # idem, met get()
```

```python
# 11.8: keys, values en items
gamescores = {"Tim": 500, "Janneke": 900, "Bob": 680}
print(gamescores.keys())    # print alle sleutels
print(gamescores.values())  # print alle waarden
print(gamescores.items())   # print alle sleutel-waarde-paren
```

```python
# 11.11: Dictionary's opslaan op je computer
import json                                  # importeer het json-pakket
gamescores = {"Tim": 500, "Janneke": 900, "Bob": 680}

uitvoer_bestand = open("score.json", "w")    # open bestand met schrijfrechten
json.dump(gamescores, uitvoer_bestand)       # schrijf dictionary naar bestand
uitvoer_bestand.close()                      # sluit het bestand
```

```python
# 11.12: Dictionary's laden van je computer
import json                                  # importeer het json-pakket

invoer_bestand = open("score.json", "r")     # open bestand met leesrechten
gamescores = json.load(invoer_bestand)       # laad dictionary uit bestand
invoer_bestand.close()                       # sluit het bestand
```
