# 13.1 Meerdere waarden retourneren met een functie
Een functie kan in Python meerdere waarden teruggeven door ze met kommaâ€™s te `returnen`.  De geretourneerde waarden sla je op in meerdere variabelen.

```python
import random

def gooi2dobbelstenen():
    worp1 = random.randint(1, 6)
    worp2 = random.randint(1, 6)
    return worp1, worp2                     # return twee waarden

dobbel1, dobbel2 = gooi2dobbelstenen()      # sla beide opgeleveerde waarden op
print(dobbel1)
print(dobbel2)
```

# 13.2 Globale variabelen gebruiken
Een globale variabele is overal in je programma te gebruiken en staat bovenaan in de code.  
Ze worden vaak gebruikt voor vaste waarden (constanten).

```python
BTW = 0.21  # 21% btw

def berekenPrijsInclBTW(prijs):
    return prijs * (1 + BTW)

prijs = berekenPrijsInclBTW(19.90)
print("Prijs inclusief btw:", prijs)
```

# 13.3 Globale variabelen aanpassen (met global)
Het aanpassen van globale variabelen is meestal niet wenselijk.  
Als het toch moet, gebruik je `global` om de globale variabele aan te passen.


```python
score = 10

def scoreAanpassen():
    global score        # gebruikt de globale variabele score (met waarde 10)
    score = 8           # past de globale waarde van score aan naar 8

scoreAanpassen()
print("Score is nu:", score)
```

# Belangrijk om te onthouden

- Globale variabelen zijn overal in je programma beschikbaar.
- Gebruik globale variabelen vooral voor vaste waarden.
- Pas globale variabelen alleen aan als het echt niet anders kan.
