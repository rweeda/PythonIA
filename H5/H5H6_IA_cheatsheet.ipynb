## Functie zonder parameters en zonder retourwaarde

```python
### FUNCTIEDEFINITIE ZONDER PARAMETERS EN ZONDER RETOURWAARDE

# Functie die een groet print
def groet():
    print("Hallo")  # print een groet

### HOOFDPROGRAMMA MET AANROEP VAN DE FUNCTIE
groet()             # roep de functie groet() aan
groet()             # roep de functie groet() aan
```

## Functie met parameters maar zonder retourwaarde


```python
### FUNCTIEDEFINITIE MET PARAMETERS MAAR ZONDER RETOURWAARDE

# Functie die een groet print.
def groet(naam, land): # de functie heeft twee parameters: naam en land
    print("Hallo", naam, "uit", land + "!")  # print een groet met de meegegeven naam en het land

### HOOFDPROGRAMMA MET AANROEP VAN DE FUNCTIE
groet("Sebas", "Spanje")      # roep de functie groet() aan met argumenten: "Sebas" als naam en "Spanje" als land
groet("John", "Engeland")
groet("Janneke", "Nederland")
```

## Functie met parameters en met retourwaarde

```python
### FUNCTIEDEFINITIE MET PARAMETERS EN MET RETOURWAARDE

# Functie die een groet print en het aantal keren dat er gegroet is bijhoudt
def groet(naam, teller):         # functie krijgt 2 parameters mee om mee te werken: naam en teller
    print("Hallo", naam)         # print een groet
    teller += 1
    return teller

### HOOFDPROGRAMMA MET AANROEP VAN DE FUNCTIE
aantal_keren = 0

aantal_keren = groet("Sebas", aantal_keren)     # roep de functie aan met "Sebas" en aantal_keren, en sla de nieuwe waarde op
aantal_keren = groet("John", aantal_keren)
aantal_keren = groet("Janneke", aantal_keren)

print("Er is", aantal_keren, "keer gegroet")
```

```python
### FUNCTIE MET PARAMETERS EN RETOURWAARDE

def kmNaarMeters(km):	  	# functie die kilometers omzet naar meters
    meters = km * 1000 		# zet kilometers om naar meters	  
    return meters			# geef het aantal meters terug

### HOOFDPROGRAMMA
aantal_meters = kmNaarMeters(4)
print(aantal_meters)
```
