## Functie zonder parameters en zonder retourwaarde

```python
### FUNCTIEDEFINITIE ZONDER PARAMETERS EN ZONDER RETOURWAARDE

# functie die een groet print
def groet( ):        
    print("Hallo")   # print een groet

### HOOFDPROGRAMMA MET AANROEP VAN FUNCTIE
groet( )            # roep de functie groet aan
groet( )            # roep de functie groet aan
```

## Functie met parameters maar zonder retourwaarde
Een functie kan informatie meekrijgen (parameters) die nodig zijn in een functie (hieronder: naam en land).
Bij de functieaanroep (regels 8 t/m 10) geef je de specifieke waarden tussen haakjes mee.
Bij de functiedefinitie gebruik je de namen zoals je ze in de functie gebruikt (regel 4).

```python
### FUNCTIEDEFINITIE MET PARAMETERS MAAR ZONDER RETOURWAARDE

 # functie die een groet print. Functie krijgt 2 parameters mee om mee te werken: naam en land
def groet( naam, land ):                       
    print("Hallo", naam, "uit", land + "!")   # print een groet met de meegegeven naam en land

### HOOFDPROGRAMMA MET AANROEP VAN FUNCTIE
groet( "Sebas", "Spanje" ) # roep de functie groet aan met "Sebas" voor naam en "Spanje" voor land
groet( "John", "Engeland" )
groet( "Janneke", "Nederland" )
```

## Functie met parameters en retourwaarde
Een functie kan informatie meekrijgen (parameters) en iets teruggeven (return).
De waarde die je terugkrijgt (hier <code>teller</code>), moet je opslaan in een variabele in het hoofdprogramma (hier: <code>aantal_keren</code>), anders ben je hem kwijt.
De naam van de variabele in het hoofdprogramma hoeft niet hetzelfde te zijn als de naam in de functie.

```python
### FUNCTIEDEFINITIE MET PARAMETERS MET RETOURWAARDE

# functie die een groet print en het aantal keren dat er gegroet is bijhoudt
def groet( naam, teller ):              # functie krijg 2 parameters mee om mee te werken: naam en land
    print( teller, "hallo", naam )      # print een groet met de meegegeven teller en naam
    teller += 1
    return teller

### HOOFDPROGRAMMA MET AANROEP VAN FUNCTIE
aantal_keren = 0
aantal_keren = groet( "Sebas", aantal_keren )   # roep de functie aan met "Sebas" en aantal_keren_geroet, en sla de nieuwe waarde op
aantal_keren = groet( "John", aantal_keren )
aantal_keren = groet( "Janneke", aantal_keren)
print("Totaal aantal keren gegroet:", aantal_keren)   # print aangepaste waarde
```

```python
### FUNCTIEDEFINITIE MET PARAMETERS MET RETOURWAARDE

# functie die gegeven kilometers omzet naar meters
def rekenKmNaarMeters( afstand_in_km ): # definitie van functie die parameter afstand_in_km binnen krijgt om mee te rekenen
	afstand_in_m = afstand_in_km * 1000	# voert berekening uit
	return( afstand_in_m ) 				# levert afstand_in_m op

### HOOFDPROGRAMMA MET AANROEP VAN FUNCTIE
aantal_meters = rekenKmNaarMeters( 4 ) #functie rekenKmNaarMeters aanroepen met 4 km als invoer, wat het oplevert opslaan als aantal_meters
print( aantal_meters )
```

```python
import turtle

### FUNCTIEDEFINITIE

# functie die een cirkel tekent met opgegeven straal en kleur
def tekenSnoep( straal, kleur ): 
  turtle.pendown()
  turtle.fillcolor( kleur) # stel vulkleur in
  turtle.begin_fill()      # begin met vullen
  turtle.circle( straal )  # teken cirkel met opgegeven straal
  turtle.end_fill()        # einde vullen

### HOOFDPROGRAMMA
ingevoerde_kleur = input("Voer de kleur in: ")        # vraag gebruiker om kleur
ingevoerde_straal = int(input("Voer de straal in: ")) # vraag gebruiker om straal
tekenSnoep( ingevoerde_straal, ingevoerde_kleur )      # roep functie aan met wensen van de gebruiker

turtle.done()
```
