## Functie met parameters maar zonder retourwaarde
Een functie kan informatie die nodig is meekrijgen als parameters in een functie (hieronder zijn de parameters <code>naam</code> en <code>land</code>).
Bij de functieaanroep (regels 8 t/m 10) geef je de specifieke waarden (argumenten) tussen haakjes mee (hieronder zijn de argumenten op regel 8 "Sebas" en "Spanje").
Bij de functiedefinitie gebruik je de namen van de parameters zoals je ze in de functie gebruikt (regel 4).

```python
### FUNCTIEDEFINITIE MET PARAMETERS MAAR ZONDER RETOURWAARDE

# Functie die een groet print.
def groet(naam, land): # de functie heeft twee parameters: naam en land
    print("Hallo", naam, "uit", land + "!")  # print een groet met de meegegeven naam en het land

### HOOFDPROGRAMMA MET AANROEP VAN DE FUNCTIE
groet("Sebas", "Spanje")      # roep de functie groet() aan met argumenten: "Sebas" als naam en "Spanje" als land
groet("John", "Engeland")
groet("Janneke", "Nederland")
```

```python
import turtle

### FUNCTIEDEFINITIE

# functie die een cirkel tekent met opgegeven straal en kleur
def tekenSnoep( straal, kleur ): 
  turtle.pendown()
  turtle.fillcolor( kleur) # stel vulkleur in
  turtle.begin_fill()      # begin met vullen
  turtle.circle( straal )  # teken cirkel met opgegeven straal
  turtle.end_fill()        # einde vullen

### HOOFDPROGRAMMA
ingevoerde_kleur = input("Voer de kleur in: ")        # vraag gebruiker om kleur
ingevoerde_straal = int(input("Voer de straal in: ")) # vraag gebruiker om straal
tekenSnoep( ingevoerde_straal, ingevoerde_kleur )      # roep functie aan met wensen van de gebruiker

turtle.done()
```

## Functie met parameters en met retourwaarde
Een functie kan informatie meekrijgen (argumenten die op de plek van de parameters terecht komen) en iets teruggeven (<code>return</code>).
De waarde die je terugkrijgt (hier: <code>teller</code>), moet je opslaan in een variabele in het hoofdprogramma (hier: <code>aantal_keren</code>), anders ben je de waarde kwijt.
De naam van de variabele in het hoofdprogramma hoeft niet dezelfde te zijn als de naam in de functie.

```python
### FUNCTIE MET PARAMETERS EN RETOURWAARDE
# functie die een groet print en het aantal keren dat er gegroet is, bijhoudt
def groet(naam, teller):    # de functie heeft twee parameters: naam en teller
    print("Hallo", naam)
    teller += 1             # verhoog de teller met 1
    return teller           # geef de nieuwe teller terug

### HOOFDPROGRAMMA
aantal_keren = 0

aantal_keren = groet("Sebas", aantal_keren) # roep de functie aan met argumenten "Sebas" en aantal_keren, en sla de nieuwe waarde op
aantal_keren = groet("John", aantal_keren)
aantal_keren = groet("Janneke", aantal_keren)

print("Er is", aantal_keren, "keer gegroet")
```

```python
### FUNCTIE MET PARAMETERS EN RETOURWAARDE

def kmNaarMeters(km):	  	# functie die kilometers omzet naar meters
    meters = km * 1000 		# zet kilometers om naar meters	  
    return meters			# geef het aantal meters terug

### HOOFDPROGRAMMA
aantal_meters = kmNaarMeters(4)
print(aantal_meters)
```
