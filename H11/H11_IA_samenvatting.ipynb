Een loop mag ook in een andere loop staan. Dit kun je gebruiken om alle combinaties van twee dingen af te drukken.

```python
# 12.5: Geneste for-loops
for x_coordinaat in range(4):
    for y_coordinaat in range(5):
        print("(", x_coordinaat, ",", y_coordinaat, ")", end="")  # end zorgt ervoor dat niet met een nieuwe regel begonnen wordt
    print()
```

```python
# 12.6: Geneste lijsten
bord = [["X", "O", "-"], ["X", "O", "O"], ["-", "X", "-"]]  # spelbord met tussenstand van een boter-kaas-en-eieren-spel

print("Linksboven:", bord[0][0])        # print element linksboven
print("Rechtsboven:", bord[0][2])       # print element rechtsboven
print("Rechtsonder:", bord[2][2])       # print element rechtsonder

# 12.7: Geneste lijst doorlopen met een geneste for-loop
for rij in range(3):                    # doorloop alle rijen van het bord
    for kolom in range(3):              # doorloop alle kolommen van het bord
        print(bord[rij][kolom], end="") # end zorgt ervoor dat niet met een nieuwe regel begonnen wordt
    print()                             # ga naar een nieuwe regel na elke rij
```
