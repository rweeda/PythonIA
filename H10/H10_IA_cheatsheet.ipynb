## 10.2 Filter-plan: nieuwe lijst met bepaalde waarden weggelaten

```python
### FUNCTIEDEFINITIE

# maakt een nieuwe lijst zonder onvoldoendes
def geefLijstZonderOnvoldoende( lijst_met_onvoldoendes ):
    lijst_zonder_onvoldoendes = []	                    # nieuwe lege lijst
    for cijfer in lijst_met_onvoldoendes:               # doorloop de gegeven lijst
        if cijfer > 5.0:                                # als het cijfer een voldoende is
            lijst_zonder_onvoldoendes.append( cijfer )  # voeg voldoende cijfer toe aan nieuwe lijst
    return lijst_zonder_onvoldoendes                    # lever de nieuwe lijst op
	
### HOOFDPROGRAMMA
cijferlijst = [8.2, 6.2, 5.0, 4.9, 7.1]
nieuwelijst = geefLijstZonderOnvoldoende( cijferlijst )
print( "Lijst zonder onvoldoendes:", nieuwelijst )
```

## 10.3 Tel-plan: aantal voorkomens tellen

```python
cijferlijst = [8.2, 6.2, 5.0, 4.9, 7.1]

onvoldoende_teller = 0				# teller begint op 0
for cijfer in cijferlijst:
	if cijfer <= 5.0:				# zodra een onvoldoende gevonden is
		onvoldoende_teller += 1		# teller ophogen
		
print( "Aantal onvoldoendes:", onvoldoende_teller )
```

## 10.4 Zoek-plan: houd met een <i>logische vlag</i> bij of element gevonden is

```python
fruitLijst  = ["banaan", "appel", "druif", "peer", "wortel"]

wortelGevonden = False			# logische vlag: er is nog geen wortel gevonden

for fruit in fruitLijst:
	if fruit == "wortel":		# als 'wortel' in de lijst is gevonden
		wortelGevonden = True	# logische vlag aanpassen


if wortelGevonden == True:
	print("Lijst doorlopen, geen 'wortel' gevonden.")
else:
	print("Lijst doorlopen, wel 'wortel' gevonden.")
```

## 10.5 Min/max-plan: kleinste of grootste waarde bepalen

```python
getallenlijst = [ 4, 2, 3, 1, 6, 7]

kleinste_tot_nu_toe = getallenlijst[0]  # pak de eerste waarde uit lijst, die is dan de kleinste tot nu toe

for getal in getallenlijst:             # doorloop alle getallen
	if getal < kleinste_tot_nu_toe:		# als er kleinere waarde wordt gevonden
		kleinste_tot_nu_toe = getal		# dan de waarde van kleinste aanpassen

print("kleinste is", kleinste_tot_nu_toe)
```

## 10.6 Som-plan: som van waarden in een lijst

```python
getallenlijst = [ 4, 2, 3, 1]

som = 0             		 # som begint op 0
for getal in getallenlijst:
	som += getal      		 # tel getal op bij de huidige som
	
print("Som is:", som)
```
