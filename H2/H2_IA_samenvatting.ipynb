# Onderwerp 2 - Interactie: Samenvatting

In dit onderwerp leer je de basis van datatypes, tekst en getallen, printen en invoer. Hieronder staan korte voorbeelden die je zelf kunt uitvoeren.


## 2.1: Datatypes: <code>string</code>, <code>int</code> en <code>float</code>

```python
naam = "Janneke"        # Een tekst noemen we een string
print(naam, type(naam))

leeftijd = 13           # Een geheel getal noemen we een int
print(leeftijd, type(leeftijd))

pi = 3.14               # Een kommagetal noemen we een float
print(pi, type(pi))
```

## 2.3&2.4: Tekst en/of getallen tonen met <code>print()</code>

```python
print("Hallo wereld")              # Je kunt één string printen
print("Voornaam:" + "Janneke")     # Met '+' plak je strings direct aan elkaar zonder spatie
print("Woonplaats:", "Madurodam")  # Met een komma scheid je meerdere strings, er komt automatisch een spatie tussen
print("Leeftijd:", 13)               # Je kunt ook andere datatypes printen, zoals integers
```

## 2.8: Getallen en tekst aan elkaar plakken met gebruik van <code>str()</code>

```python
uren = 3            # uren is een getal (int)
minuten = 21        # minuten is een getal (int)
tijd = "De tijd is " + str(uren) + ":" + str(minuten)  # maak met str() een string van het getallen uren en minuten, en plak alles aan elkaar
print(tijd)
```

## 2.9: Invoer vragen met <code>input()</code>

```python
naam = input("Wat is je naam? ") # met input() kun je de gebruiker om invoer vragen
print("Hallo, " + naam)          # standaard is de invoer een string (tekst)
```

## 2.10: Een getal als invoer vragen met <code>input()</code> en omzetten naar <code>int</code>

```python
invoer = input("Geef een geheel getal: ")    # standaard is de invoer een string (tekst)
getal_int = int(invoer)                      # zet invoer om naar int (getal)

print("Je gaf het geheel getal:", getal_int)    
```

## 2.10: Een kommagetal als invoer vragen met <code>input()</code> en omzetten naar <code>float</code>

```python
invoer = input("Geef een kommagetal: ")    # standaard is de invoer een string (tekst)
kommagetal = float(invoer)                 # zet invoer om naar float (kommagetal) 

print("Je gaf het kommagetal:", kommagetal)
```

## 2.11: Rekenen met getallen

```python
a = 10
b = 3
print("Optellen:", a + b)
print("Aftrekken:", a - b)
print("Vermenigvuldigen:", a * b)
print("Delen:", a / b)
print("Machtsverheffen:", a ** b)   # a tot de macht b
print("Modulo (rest):", a % b)      # klokrekenen: steeds b eraf halen
```
