# Onderwerp 7 - <code>while</code> ('<i>zolang</i>'): Cheatsheet



```python
# Print alle even getallen tussen 0 en 10
getal = 0
while getal < 10:      # zolang getal kleiner is dan 10
    print(getal)
    getal += 2         # verhoog getal met 2

print("Klaar!")
```

## <code>while</code> met een <i>logische vlag</i>

```python
WACHTWOORD = "geheim"
goed_geraden = False                                     # om te beginnen staat de vlag op False

while not goed_geraden:                                  # herhaal zolang nog niet goed geraden
    invoer = input("Geef je wachtwoord op: ")            # vraag gebruiker om wachtwoord

    if invoer == WACHTWOORD:                             # als er goed is geraden
        print("Goed geraden!")
        goed_geraden = True                              # zet de vlag op True; hierna stopt de while-loop
    else:                                                # er is fout geraden
        print("Jammer, fout... probeer het nog een keer")# de vlag blijft False, de loop wordt herhaald

print("Einde: wachtwoord is goed geraden!")              # na de while-loop weet je zeker dat het wachtwoord goed is geraden
```
