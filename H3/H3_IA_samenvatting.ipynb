## 3.1 Waarde toekennen aan een variabele

```python
leeftijd = 18     # variabele leeftijd krijgt de waarde 18 met =
print(leeftijd)  # toont 18
```

## 3.2: Waarde van een variabele veranderen

```python
score = 10          # startwaarde is 10
score = score + 1   # verhoog score met 1
print(score)
```

## 3.7: Verkorte operatoren: +=, -=, *=, /=

```python
n = 80   # variabele n krijgt de waarde 80  
n += 5   # 5 bij n optellen, zelfde als: n = n + 5
print(n)

n -= 3   # 3 van n aftrekken, zelfde als: n = n - 3
print(n)

n *= 2   # n met 2 vermenigvuldigen, zelfde als: n = n * 2
print(n)

n /= 4   # n delen door 4, zelfde als: n = n / 4
print(n)
```

## 3.8: Standaardfuncties <code>max()</code>, <code>min()</code>, <code>pow()</code>, <code>round()</code>, <code>abs()</code>, <code>len()</code>

```python
getallen = [3, 10, -2, 7]   # lijst met getallen
print(max(getallen))   # grootste waarde kiezen
print(min(getallen))   # kleinste waarde kiezen
print(pow(2, 3))       # 2 tot de macht 3 (dus 2*2*2)
print(round(3.65, 1))  # afronden van 3.65 op 1 plaats achter de komma
print(abs(-8))         # absolute waarde: maak van een negatief getal een positief getal

woord = "informatica"
print(len(woord))      # telt het aantal tekens (lengte van de string)
```

## 3.9: Willekeurig getal: <code>randint()</code> en <code>random</code>

```python
import random                               # importeer de module random

dobbelsteen = random.randint(1, 6)          # kiest een willekeurig geheel getal uit 1 t/m 6 (zoals een dobbelsteen)
print(dobbelsteen)

willekeurig_kommagetal = random.random()    # kiest een willekeurig kommagetal tussen 0 (incl.) en 1.0 (excl.)
print(willekeurig_kommagetal)
```
