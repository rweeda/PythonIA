<p>In het schema hieronder zie je dat er een relatie is getekend tussen <b>boncode</b> uit tabel <i>bestellingen</i> en <b>boncode</b> uit tabel <i>kortingsbonnen</i>.</p>

<p><img
src="https://raw.githubusercontent.com/rweeda/PythonIA/main/sql/img/tabel_kortingsbonnen_gekoppeld_bestelling.png"
alt="Tabel kortingsbonnen" width="500"></p>



<p>De relatie tussen tabellen leg je vast met een <b>foreign key</b>. Een <i>foreign key</i> is een kolom in een tabel die verwijst naar de <i>primary key</i> van een andere tabel. In dit voorbeeld verwijst de foreign key <b>boncode</b> uit tabel <i>bestelling</i> naar de primary key <b>boncode</b> uit tabel <i>kortingsbonnen</i>.</p>

<hr>
Samengevat:
    <ul>
      <li>Een <b>foreign key</b> is een kolom in een tabel die verwijst naar de <b>primary key</b> van een andere tabel.</li>
      <li>Het doel is om gegevens uit twee tabellen logisch met elkaar te verbinden.</li>
      <li>De kolommen die aan elkaar gekoppeld zijn (bijvoorbeeld <b>boncode</b>) moeten dezelfde soort gegevens bevatten.</li>
      <li>Door deze koppeling kun je bij elke bestelling direct de juiste kortingsgegevens opzoeken.</li>
    </ul>
