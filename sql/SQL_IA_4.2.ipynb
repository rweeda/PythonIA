
## 4.2: Aantal rijen beperken met <code>LIMIT</code>
<p>Met <code>LIMIT</code> kun je bepalen hoeveel rijen er getoond moeten worden.</p>

```SQL
SELECT kolomnaam
FROM tabel
LIMIT aantal_rijen;
```
