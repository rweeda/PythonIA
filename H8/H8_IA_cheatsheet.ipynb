# Onderwerp 8 - <code>for</code>-loop: Cheatsheet


##  1 parameter: eindgrens

```python
# Herhaalt de ingesprongen code 3 keer: voor teller 0 tot 3 (niet tot en met 3)
for teller in range(3):
    print(teller, "Hallo")

print("Klaar!")
```

## 2 parameters: startgetal en eindgrens

```python
# Herhaalt de ingesprongen code 3 keer: voor teller van 2 tot 5 (niet tot en met 5)
for teller in range(2, 5):
    print(teller)

print("Klaar!")
```

## 3 parameters: startgetal, eindgrens en stapgrootte


```python
# Herhaalt de ingesprongen code voor getal van 13 tot 2 (niet tot en met 2), met stapgrootte -3
for getal in range(13, 2, -3):
    print(getal)

print("Klaar!")
```
