# Onderwerp 8 - <code>for</code>-loop: Samenvatting

<p>Met een <code>for</code>-loop kun je code een aantal keren herhalen.</p>


## 8.1: <code>for in range()</code> met 1 parameter: eindgrens

```python
# herhaalt de ingesprongen code 3 keer: voor teller 0 tot 3 (niet tot en met 3)
for teller in range (3):
	print(teller, "Hallo")
print("Klaar!")
```

## 8.3: <code>for in range()</code> met 2 parameters: startgetal en eindgrens

```python
# herhaalt de ingesprongen code 3 keer: voor teller van 2 tot 5 (niet tot en met 5)
for teller in range( 2, 5 ):
  print( teller )
print( "Klaar!" )
```

## 8.3: <code>for in range()</code> met 3 parameters: startgetal, eindgrens en stapgrootte


```python
# herhaalt de ingesprongen code voor getal van 13 tot 2 (niet tot en met 2), met stapgrootte -3
for getal in range( 13, 2, -3 ):
	print( getal )
print( "Klaar!" )
```
