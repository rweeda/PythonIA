## 2. Algoritmen

Elke app of elk programma heeft een doel: routes plannen, treintijden tonen, socialmediaposts weergeven of een spelletje spelen. Deze software wordt geprogrammeerd. Voor een goed resultaat bepaal je vooraf wat het programma precies doet en hoe je dat bereikt.

Dat <i>hoe</i> leg je vast in een <b>algoritme</b>. Dit is een stappenplan met instructies om een taak uit te voeren. Dat kan een complex systeem zijn, maar ook een eenvoudige berekening zoals de gemiddelde dagtemperatuur. Een goed algoritme schrijven is uitdagend: je mag geen details vergeten en de volgorde moet exact kloppen.

<b>Invoer-Verwerking-Uitvoer</b>
<p>Een algoritme is een stappenplan of lijst instructies om een taak uit te voeren. Een algoritme krijgt <b>invoer</b> die het nodig heeft om de taak uit te voeren, en levert een resultaat: de <b>uitvoer</b>. Vergelijk het met een recept voor poffertje: daar staan stappen in zoals “Mix de bloem, eieren en melk”, “Giet een beetje beslag in de kuiltjes van de poffertjespan” en “Keer de poffertjes om”. Tijdens het bakken voer je de instructies stap voor stap uit. De invoer bestaat uit de bloem, de eieren en melk; de uitvoer is de gebakken poffertjes.</p>


<p><img src="https://raw.githubusercontent.com/rweeda/PythonIA/main/img/poffertjes.jpg" width="50%" height="50%" alt="Programmeren net als het bakken van een poffertjes met Invoer - Verwerking - Uitvoer">

Hieronder zie je drie voorbeelden van verschillende soorten invoer, verwerking en uitvoer:</p>


<p><img src="https://raw.githubusercontent.com/rweeda/PythonIA/main/img/Py2_invoer_verwerking_uitvoer.png" width="50%" height="50%" alt="Gebruikersinvoer bij verschillende digitale systemen" role="presentation" class="atto_image_button_right">

<ul>
<li>Voorbeeld 1: Je geeft aan met hoeveel personen je naar welke film wilt gaan (invoer). Het te betalen bedrag wordt berekend (verwerking). Daarna verschijnt het bedrag op het scherm (uitvoer).</li>

<li>Voorbeeld 2: Je voert je wachtwoord in (invoer). Het systeem controleert of je toegang hebt (verwerking) en opent de app (uitvoer).</li>

<li>Voorbeeld 3: Je drukt op de knop bij een verkeerslicht (invoer). De computer bepaalt het moment van oversteken (verwerking) en vervolgens gaan de andere lichten op rood en jouw licht op groen (uitvoer).</li>
</ul>

Kortom: <i>invoer</i> is alle informatie die een digitaal systeem ontvangt. Invoer kan op verschillende manieren binnenkomen, bijvoorbeeld doordat je tekst intypt op een toetsenbord, een knop indrukt, een sensor iets meet of een pasje wordt gelezen. Bij de <i>verwerking</i> gebeurt er iets met die invoer: het systeem rekent iets uit, vergelijkt gegevens of controleert of aan een voorwaarde is voldaan. De <i>uitvoer</i> is het resultaat van die verwerking. Dat kan zijn dat er informatie op een scherm verschijnt, dat een poort opengaat, of dat een geluid wordt afgespeeld.

## 4. Van algoritme naar computerinstructies

Een computer is eigenlijk niet zo slim. Hij bestaat uit heel veel kleine schakelaars die alleen aan of uit kunnen staan. Deze standen worden weergegeven met 1 (aan) en 0 (uit). Daarom “begrijpt” een computer alleen reeksen van 0-en en 1-en. Dit noemen we <i>binaire code</i>.
Voor één programma zijn vaak miljarden 0-en en 1-en nodig. Die binaire code is voor mensen moeilijk te lezen en te schrijven. Daarom gebruiken we <i>programmeertalen</i> zoals Python. Dit zijn talen die voor mensen begrijpelijk zijn, maar wel strenge regels hebben, zodat de computer precies weet wat de bedoeling is.
Wanneer je op 'Run' klikt, wordt de Python‑code door een <i>interpreter</i> stap voor stap vertaald naar binaire instructies die de computer kan uitvoeren. Je kunt dit vergelijken met een tolk die live vertaalt van mensentaal naar de taal die de computer begrijpt.

## 10. Een figuur inkleuren

Om een figuur in te kleuren moet je eerst een kleur aangeven en daarna aangeven waar de code voor het figuur begint en eindigd. Zie het voorbeeld hieronder:
<ul>
<li>Kleur aangeven: <code>turtle.fillcolor("red")</code> (regel 3);</li>
<li>Begin van figuur aangeven: <code>turtle.begin_fill()</code> (regel 5);</li>
<li>Einde van figuur aangeven: <code>turtle.end_fill()</code> (regel 7);</li>
</ul>


```python
import turtle           # aangeven dat we de turtle gaan gebruiken

turtle.fillcolor("red") # kleur van figuur

turtle.begin_fill()     # hier begint figuur
turtle.circle(50)
turtle.end_fill()       # hier eindigt figuur

turtle.done()           # klaar met tekenen, maar sluit het tekenvenster niet meteen af
```

### Opdracht 10.1: Oranje driehoek

De code hieronder tekent een driehoek. Pas de code aan zodat Turtle de driehoek oranje (orange) inkleurt.

```python
import turtle       # gebruik turtle-module

#BEGIN MET TEKENEN VAN DRIEHOEK
turtle.forward(20)          # loop 20 stappen vooruit
turtle.left(120)            # draai 120 graden naar links
turtle.forward(20)          # loop 20 stappen vooruit
turtle.left(120)            # draai 120 graden naar links
turtle.forward(20)          # loop 20 stappen vooruit
#KLAAR MET TEKENEN DRIEHOEK

turtle.done()               # klaar met tekenen, maar sluit het tekenvenster niet meteen af
```

```python
# ANTWOORD Oranje driehoek

import turtle               # importeer turtle-module

turtle.fillcolor("orange")  # Kleur van de pen. Je kunt ook kiezen uit black, tan, gold, blue, red, ...

turtle.begin_fill()         # begin van het figuur dat ingekleurd moet worden
#BEGIN MET TEKENEN VAN DRIEHOEK
turtle.forward(20)          # loop 20 stappen vooruit
turtle.left(120)            # draai 120 graden naar links
turtle.forward(20)          # loop 20 stappen vooruit
turtle.left(120)            # draai 120 graden naar links
turtle.forward(20)          # loop 20 stappen vooruit
#KLAAR MET TEKENEN DRIEHOEK
turtle.end_fill()           # eind van het figuur dat ingekleurd moet worden

turtle.done()               # klaar met tekenen, maar sluit het tekenvenster niet meteen af
```

## 11. Handige Turtle-functies

<p class="cursor-hover">Een paar handige Turtle-functies zijn:</p>

```python
import turtle  # importeer turtle-module, code altijd hiermee beginnen


turtle.pencolor("darkViolet") # kleur van de pen. Je kunt ook kiezen uit black, tan, gold, blue, red, ...

turtle.shape("turtle")      # maakt een schildpadje van de cursor in plaats van een pijltje, kan ook "arrow", "triangle", "circle", "square" zijn

turtle.right(90)            # draai 90 graden naar rechts
turtle.left(45)             # draai 45 graden naar links
turtle.forward(10)          # loop 10 stappen vooruit

turtle.goto(40, 40)         # zet de schildpad op positie (x,y). (0,0) is het midden van het scherm
turtle.setheading(90)       # kijkrichting. 90 is naar boven, 0 is naar rechts

turtle.fillcolor("red")     # kleur om figuur mee in te kleuren
turtle.begin_fill()         # begin van het figuur dat ingekleurd moet worden
turtle.end_fill()           # eind van het figuur dat ingekleurd moet worden

turtle.pendown()            # zet pen op papier, bewegingen hierna zijn zichtbaar
turtle.penup()              # haal pen van papier, bewegingen hierna zijn onzichtbaar


turtle.circle(50)           # teken een cirkel met straal 50
turtle.circle(50, 180)      # teken een halve cirkel (180 graden) met straal 50
turtle.pensize(8)           # dikte van de pen
turtle.write("Hallo")       # schrijf tekst op het scherm

turtle.done()               # altijd achteraan toevoegen: klaar met tekenen, maar sluit het tekenvenster niet af
```

<p>Als je begint, staat Turtle altijd in het midden op positie (0,0), kijkend naar rechts.</p>
