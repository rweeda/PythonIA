```python
# 16.1 Geneste dictionary maken
steden = {
    "Amsterdam":{"land": "Nederland", "inwoners": 921000},
    "Parijs": {"land": "Frankrijk", "inwoners": 2148000},
    "Berlijn": {"land": "Duitsland", "inwoners": 3769000}
}

#16.2 Gegevens opvragen
print(steden)                           # Print de hele dictionary van steden
print(steden["Parijs"])                 # Print de dictionary voor Parijs, dus: {'land': 'Frankrijk', 'inwoners': 2148000}
print(steden["Parijs"]["inwoners"])     # Print aantal inwoners van parijs, dus: 2148000
print(steden["Parijs"]["land"])         # Print land van parijs, dus: Frankrijk


#16.3 Gegevens aanpassen
steden["Berlijn"]["inwoners"] = 3800000     # Pas het aantal inwoners van Berlijn aan

#16.4 Dictionary toevoegen  
steden["Madrid"] =  {"land": "Spanje", "inwoners": 3400000}  # Voeg Madrid toe aan de dictionary van steden

#16.5 Dictionary verwijderen
del steden["Berlijn"]                   # Verwijder de stad Berlijn (sleutel) uit de dictionary
print(steden)                           # Print de dictionary

#16.6 Aantal gegevens in een dictionary tellen
print("Aantal steden:", len(steden))          # Print het aantal steden in de dictionary

#16.7: Geneste dictionary doorlopen met een <code>for</code>-loop
for stad, gegevens in steden.items():   # Door de steden lopen 
    print( stad,  "ligt in", gegevens['land'], "en heeft", gegevens['inwoners'], "inwoners.") # alle gegevens printen

#16.8 Zoeken met logische vlag
gevonden = False	#logische vlag: waarde nu nog niet gevonden

for stad_info in steden.values(): # loop door de waardes van de dictionary
    if stad_info["inwoners"] < 1000000:  # als de waarde aan de voorwaarde voldoet
        gevonden = True # zet de vlag op True

if gevonden == True: # als de vlag op True staat
	print("Stad met minder dan 1 miljoen inwoners gevonden!") # print bevestiging dat waarde is gevonden
else:
	print("Geen stad met minder dan 1 miljoen inwoners gevonden.") # of print dat de waarde niet is gevonden
      

#16.9: Waarden filteren en in een nieuwe dictionary zetten
zuid_amerika_dictionary = {}                        # maak een lege dictionary aan voor Zuid-Amerikaanse steden
for stad, gegevens in steden.items():               # loop door de steden dictionary en haal de gegevens voor elke stad op
	if gegevens["werelddeel"] == "Zuid-Amerika":    # als de werelddeel Zuid-Amerika is
		zuid_amerika_dictionary[stad] = gegevens	# voeg stad toe aan zuid_amerika_dictionary

print("Dictionary met Zuid-Amerikaanse steden:", zuid_amerika_dictionary)

#16.10: Bepaalde gegevens tellen
aantalMiljoenensteden = 0	#teller begint op 0
for stad in steden:
	if steden[stad]["inwoners"] > 1000000:	#als de inwoners meer zijn dan 1 miljoen
		aantalMiljoenensteden += 1	#teller ophogen zodra element gevonden is

print("Aantal miljoenensteden is:", aantalMiljoenensteden)


#16.11 en 16.12: Logische opertoren AND, OR en NOT
for stad, gegevens in steden.items():	#voor elke stad in de dictionary steden
    # AND
    if steden[stad]["inwoners"] > 1000000 and steden[stad]["werelddeel"]=="Europa": # steden met meer dan 1 miljoen inwoners EN in Europa
        print(stad, "heeft meer dan 1 miljoen inwoners EN ligt in Europa.")
    
    # OR
    if steden[stad]["inwoners"] > 1000000 or steden[stad]["werelddeel"]=="Europa":	# steden met meer dan 1 miljoen inwoners OF in Europa
        print(stad, "heeft meer dan 1 miljoen inwoners OF ligt in Europa.")

    # NOT        
    if not steden[stad]["werelddeel"]=="Europa":	# steden die NIET in Europa liggen
        print(stad, "ligt NIET in Europa.")

    # haakjes bij combinaties van AND, OR en NOT
    if (steden[stad]["inwoners"] < 1000000 and steden[stad]["werelddeel"]=="Europa") or steden[stad]["werelddeel"]=="Azië":	# steden met (minder dan 1 miljoen inwoners EN in Europa), OF de steden in Azië
        print("(",stad, "heeft minder dan 1 miljoen inwoners EN ligt in Europa ) OF ligt in Azië.")


#16.13: Som van waarden
som = 0                                         # startwaarde van som is 0
for stad in steden:                             # doorloop alle steden in de dictionary
	som += steden[stad]["inwoners"]             # tel het aantal inwoners van elke stad op bij som
print("Het totaal aantal inwoners is:", som)    # print het totaal aantal inwoners

#16.14: Gemiddelde van waarden (gebruikmakend van som uit 16.13)
print("Het aantal steden is:", len(steden))                         # bepaal het aantal steden in de dictionary
gemiddelde = som / len(steden)                                      # bereken het gemiddelde
print("Het gemiddelde aantal inwoners per stad is:", gemiddelde)    # print het gemiddelde

#16.15: Kleinste of grootste waarde bepalen
kleinste_waarde = float("inf")                         # begin met oneindig groot (als je grootste zoekt, begin je met oneindig klein: -inf)

for stad in steden:                                    # doorloop alle steden in de dictionary
    if steden[stad]["inwoners"] < kleinste_waarde:     # Vergelijk het aantal inwoners met het huidige minimum
        kleinste_waarde = steden[stad]["inwoners"] 	 	

print("De kleinste stad heeft", kleinste_waarde, "inwoners.")
```
