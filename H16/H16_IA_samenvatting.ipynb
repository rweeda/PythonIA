```python
# 16.1 Geneste dictionary maken
steden = {
    "Amsterdam": {"land": "Nederland", "inwoners": 921000},
    "Parijs": {"land": "Frankrijk", "inwoners": 2148000},
    "Berlijn": {"land": "Duitsland", "inwoners": 3769000}
}

# 16.2 Gegevens opvragen
print(steden)                           # print de hele dictionary van steden
print(steden["Parijs"])                 # print de dictionary van Parijs
print(steden["Parijs"]["inwoners"])     # print het aantal inwoners van Parijs
print(steden["Parijs"]["land"])         # print het land van Parijs


# 16.3 Gegevens aanpassen
steden["Berlijn"]["inwoners"] = 3800000 # pas het aantal inwoners van Berlijn aan


# 16.4 Dictionary toevoegen
steden["Madrid"] = {"land": "Spanje", "inwoners": 3400000}  # voeg Madrid toe


# 16.5 Dictionary verwijderen
del steden["Berlijn"]                   # verwijder Berlijn uit de dictionary
print(steden)


# 16.6 Aantal gegevens in een dictionary tellen
print("Aantal steden:", len(steden))


# 16.7 Geneste dictionary doorlopen met een for-loop
for stad, gegevens in steden.items():
    print(stad, "ligt in", gegevens["land"], "en heeft", gegevens["inwoners"], "inwoners.")


# 16.8 Zoeken met een logische vlag
gevonden = False                        # logische vlag: nog niets gevonden

for stad_info in steden.values():
    if stad_info["inwoners"] < 1000000: # steden met minder dan 1 miljoen
        gevonden = True

if gevonden:
    print("Stad met minder dan 1 miljoen inwoners gevonden!")
else:
    print("Geen stad met minder dan 1 miljoen inwoners gevonden.")
```

```python
steden = {
    "Amsterdam": {"land": "Nederland", "werelddeel": "Europa", "inwoners": 921000},
    "Parijs": {"land": "Frankrijk", "werelddeel": "Europa", "inwoners": 2148000},
    "Berlijn": {"land": "Duitsland", "werelddeel": "Europa", "inwoners": 3769000},
    "Bogota": {"land": "Colombia", "werelddeel": "Zuid-Amerika", "inwoners": 7744000},
    "Tokio": {"land": "Japan", "werelddeel": "Azië", "inwoners": 13960000}
}

# 16.9 Waarden filteren en in een nieuwe dictionary zetten
zuid_amerika_dictionary = {}            # lege dictionary voor Zuid-Amerikaanse steden

for stad, gegevens in steden.items():
    if gegevens["werelddeel"] == "Zuid-Amerika":
        zuid_amerika_dictionary[stad] = gegevens

print("Dictionary met Zuid-Amerikaanse steden:", zuid_amerika_dictionary)


# 16.10 Bepaalde gegevens tellen
aantal_miljoenensteden = 0              # teller begint op 0

for stad in steden:
    if steden[stad]["inwoners"] > 1000000:
        aantal_miljoenensteden += 1

print("Aantal miljoenensteden is:", aantal_miljoenensteden)


# 16.11 en 16.12 Logische operatoren AND, OR en NOT
for stad, gegevens in steden.items():

    # AND
    if gegevens["inwoners"] > 1000000 and gegevens["werelddeel"] == "Europa":
        print(stad, "heeft meer dan 1 miljoen inwoners EN ligt in Europa.")

    # OR
    if gegevens["inwoners"] > 1000000 or gegevens["werelddeel"] == "Europa":
        print(stad, "heeft meer dan 1 miljoen inwoners OF ligt in Europa.")

    # NOT
    if not gegevens["werelddeel"] == "Europa":
        print(stad, "ligt NIET in Europa.")

    # combinaties met haakjes
    if (gegevens["inwoners"] < 1000000 and gegevens["werelddeel"] == "Europa") or gegevens["werelddeel"] == "Azië":
        print("(", stad, "heeft minder dan 1 miljoen inwoners EN ligt in Europa ) OF ligt in Azië.")


# 16.13 Som van waarden
som = 0
for stad in steden:
    som += steden[stad]["inwoners"]

print("Het totaal aantal inwoners is:", som)


# 16.14 Gemiddelde van waarden
print("Het aantal steden is:", len(steden))
gemiddelde = som / len(steden)
print("Het gemiddelde aantal inwoners per stad is:", gemiddelde)


# 16.15 Kleinste waarde bepalen
kleinste_waarde = float("inf")           # start met oneindig groot

for stad in steden:
    if steden[stad]["inwoners"] < kleinste_waarde:
        kleinste_waarde = steden[stad]["inwoners"]

print("De kleinste stad heeft", kleinste_waarde, "inwoners.")
```
