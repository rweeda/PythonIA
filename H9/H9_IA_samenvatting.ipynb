```python
boodschappenlijst = [ "appels", "brood", "melk" , "kaas"]	# maak een boodschappenlijst
print( boodschappenlijst )                                  # print de boodschappenlijst

op_index2 = boodschappenlijst[2]                            # element op index 2 (het 3e element)
print( "Op index 2 staat:", op_index2 )                    

lengte = len(boodschappenlijst)                             # lengte (aantal elementen)
print( "aantal elementen in lijst:", lengte)

achterste_element = boodschappenlijst[ lengte - 1 ]         # haal het laatste element op
print( "Het laatste element is:", achterste_element )       # druk het laatste element af


index_van_melk = boodschappenlijst.index( "melk" )          # bepaal de index van een element
print( "De index van melk is:", index_van_melk )            # druk de index van melk af

print("Melk in lijst?", 'melk' in boodschappenlijst)        # controleer of 'melk' in de lijst zit
print("Chocola in lijst?", 'chocola' in boodschappenlijst)  # controleer of 'chocola' in de lijst zit

boodschappenlijst[2] = "cola"                               # wijzig element op een bepaalde index
print( boodschappenlijst )

snoeplijst = [ "drop", "chocolade", "kauwgom" ]             # maak een snoeplijst
nieuwe_boodschappenlijst = boodschappenlijst + snoeplijst   # voeg lijsten samen
print( nieuwe_boodschappenlijst )                        

nieuwe_boodschappenlijst.remove( "kaas" )                   # verwijder uit de lijst
nieuwe_boodschappenlijst.append( "chips" )                  # voeg achteraan toe
nieuwe_boodschappenlijst.insert( 2, "yoghurt" )             # voeg toe op een bepaalde index
print( nieuwe_boodschappenlijst )

print("doorloop elementen in lijst:")   
for eten in nieuwe_boodschappenlijst:                       # doorloop alle elementen in de lijst
    print( eten )                                           # druk het element af
```

## 9.13 Willekeurig element kiezen

```python
import random                                            # importeer de random module

kleurenlijst = ["rood", "blauw", "groen", "geel"]
willekeurig_kleur = random.choice( kleurenlijst )       # kies een willekeurig element uit de lijst

print( "Willekeurig gekozen kleur is:", willekeurig_kleur )
```
