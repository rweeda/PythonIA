# Onderwerp 13 - Strings: Samenvatting



```python
# 13.1 Strings (herhaling)

woord = "fiets"
woord2 = 'bel'
print( woord )
print( woord + woord2 ) # plak twee strings aan elkaar
print( 3 * woord )     # herhaal string 3 keer
print( len(woord) )    # lengte van string


if "e" in woord:       # controleer of letter 'e' in woord1 zit
	print( "e gevonden in "+ woord )
	

# 13.2 Strings over meerdere regels
lange_string = """Python is een programmeertaal die begin jaren 90 ontworpen en ontwikkeld werd door Guido van Rossum.
        Python is ontwikkeld met het oog op leesbare code.
        (Bron: Wikipedia)"""

#13.3 Nieuwe regel in een string met '\n'
print("Hoi\nHallo") # nieuwe regel tussen Hoi en Hallo

#13.4 Index: positie van karakters in een string
print( woord[0] )                # print letter op index 0, dus de 1e letter: 'f'
print( woord[3] )                # print letter op index 3, dus de 5e letter: 't'
print( woord[ len(woord)-1 ] )   # print laatste letter van woord: 's'

#13.5 Slicing: strings opknippen
deel_van_woord = woord[1:4]  # indices 1 tot 4 (oftewel index 1 t/m 3): 'iet'
print(deel_van_woord)

#13.6 Strings doorlopen met een for-loop
for letter in woord:        # voor elke letter in woord
    print("hudige letter is:", letter)


#13.7 Strings doorlopen met een while-loop
index = 0                       # begin bij de eerste letter
while index < len( woord ):     # zolang de index kleiner is dan de lengte van het woord
    print( woord[index] )       # print de letter op de huidige index
    index += 1                  # verhoog de index met 1 voode de volgende letter

#13.8 Handige string functies
woord_met_spaties = "  fiets  "
woord_zonder_spaties = woord_met_spaties.strip()   # verwijderst spaties aan begin en eind
print("[start]" + woord_zonder_spaties + "[einde]")

hoofdletters = woord.upper()    # maakt alle letters hoofdletters
kleineletters = woord.lower()    # maakt alle letters kleine letters

woord.find("ie")          # zoekt naar substring "ie" in woord, geeft index terug waar substring begint
woord.replace("ie", "ee")  # vervangt "ie" door "ee" in woord
```
