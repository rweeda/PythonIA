# Onderwerp 15 – Tekstbestanden: Samenvatting

<ul>
    <li>Een regeleinde (ENTER) wordt gecodeerd als <code>\n</code>.</li>
    <li>Een bestand moet je altijd eerst openen met <code>open()</code>. Je geeft aan wat
        je van plan bent te doen met:
        <code>"r"</code> (read: lezen),
        <code>"w"</code> (write: schrijven),
        <code>"a"</code> (append: achteraan toevoegen),
        of <code>"w+"</code> / <code>"a+"</code>.
        <ul>
            <li><code>r</code>: alleen lezen uit het bestand</li>
            <li><code>w</code>: schrijven in het bestand. Bestaat het bestand nog niet, dan wordt het aangemaakt.
                Let op: hiermee overschrijf je de bestaande inhoud.</li>
            <li><code>w+</code>: lezen en schrijven. Bestaat het bestand nog niet, dan wordt het aangemaakt.</li>
            <li><code>a</code>: achteraan aan het bestand toevoegen. Bestaat het bestand nog niet, dan wordt het aangemaakt.</li>
            <li><code>a+</code>: lezen en achteraan schrijven. Bestaat het bestand nog niet, dan wordt het aangemaakt.</li>
        </ul>
    </li>
    <li><code>read()</code> leest <b>alle inhoud</b> in als één <code>string</code>.</li>
    <li><code>readline()</code> leest <b>één regel</b> in als <code>string</code>.</li>
    <li><code>readlines()</code> leest <b>alle regels</b> in en zet ze in een <b>lijst</b>.</li>
    <li>Met <code>.split(' ')</code> zet je een string om in een lijst. De scheiding gebeurt
        op basis van het opgegeven teken.</li>
    <li><code>write()</code>, <code>writeline()</code> en <code>writelines()</code>
        gebruik je om naar een bestand te schrijven.</li>
    <li>Een bestand sluit je af met <code>close()</code>.</li>
</ul>

```python
#15.2 Openen, lezen, printen en sluiten

bestand = open( "groet.txt", "r" )   # open het bestand om te lezen ("r")
tekst_uit_bestand = bestand.read()   # lees de inhoud van het bestand en sla het op in een variabele
print(tekst_uit_bestand)             # print naar de console
bestand.close()                      # sluit het bestand

```

```python
#15.3 Een of meerdere regels lezen

# EEN regel lezen met readline()

bestand = open("dikkertjeDap.txt", "r") # open het bestand om te lezen ("r")

regel = bestand.readline()              # lees eerste regel
print(regel)                            

regel = bestand.readline()              # lees tweede regel
print(regel)

# ALLE regels in een lijst met readlines()
inhoud_in_lijst = bestand.readlines() #zet inhoud in een lijst, elk regel apart

bestand.close()             # sluit het bestand
```

```python
#15.4 In een bestand schrijven met write()
bestand = open("ingevoerdeTekst.txt", "w")  # open het bestand om te schrijven ("w")
bestand.write(" Hallo" )                    # schrijf een STRING naar het bestand")
bestand.close()                           

#15.5 Een lijst in een bestand scrhijven met writelines()
bestand = open("uitvoer.txt", "w")          # open het bestand om te schrijven ("w")
bestand.writelines( ["hoi\n","hallo\n"] )   # schrijf de LIJST naar het bestand, maar voegt zelf GEEN '\n' toe
bestand.close()

#15.6 Toevoegen achteraan een bestand met append()
bestand_handle = open("uitvoer.txt", "a")   # open het bestand om toe te voegen ("a")
bestand_handle.write( "Hallo" )
bestand_handle.close()
```

```python
#15.7 CSV-bestanden lezen en schrijven
# Bij CSV zijn waarden gescheiden door komma's

bestand = open("stand.csv", "a")            # 'a' om achteraan toe te voegen

# Voeg score van Bob toe: 300
bestand.write("\n")                          # nieuwe regel
bestand.write("Bob")
bestand.write(",")                           # komma als scheidingsteken
bestand.write(str(300))                     # getallen omzetten naar tekst

bestand.close()
```
